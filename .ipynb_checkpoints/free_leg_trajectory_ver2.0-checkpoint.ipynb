```julia



#support time 0.5
#t0=0.5
#te=1.0
t0=100
te=200

#xの遊脚軌道

#条件ベクトル
x0=0.2
xe=0.3
v0=-0.2
ve=-0.2
ae=-0.1

#A=[x0; xe; v0; ve; ae]

#B=[t0^4 t0^3 t0^2 t0 1;
#    te^4 te^3 te^2 te 1;
#    4t0^3 3t0^2 2t0 1 0;
#    4te^3 3te^2 2te 1 0;
#    12te^2 6te 2 0 0]
A=[x0; xe; v0; ve]

B=[t0^3 t0^2 t0 1
    te^3 te^2 te 1
    3t0^2 2t0 1 0
    3te^2 2te 1 0]

C=inv(B)*A

dt=0.01 #離散時間
t=[i for i in t0:dt:te]; #描画時間
#x=C[1]*.^(t,4)+C[2]*.^(t,3)+C[3]*.^(t,2)+C[4]*t+C[5] #位置
#v=4*C[1]*.^(t,3)+3*C[2]*.^(t,2)+2*C[3]*t+C[4] #速度
#a=12*C[1]*.^(t,2)+6*C[2]*t+2*C[3] #加速度
x=C[1]*.^(t,3)+C[2]*.^(t,2)+C[3]*t+C[4] #位置
v=3*C[1]*.^(t,2)+2*C[2]*t+C[3] #速度
a=6*C[1]*t+2*C[2]
```

```julia
plot(t,x,st=:scatter);
```

```julia
plot(t,v,st=:scatter);
```

```julia
plot(t,a,st=:scatter);
```

```julia
#zの遊脚軌道
zh=0.05
z0=0.0
ze=0.0
v0=0.0
ve=0.0

#support time 0.5
t0=0.5
te=1.0
th=(te-t0)/2+t0

A_z=[zh; z0; ze; v0; ve]

B_z=[th^4 th^3 th^2 th 1
    t0^4 t0^3 t0^2 t0 1
    te^4 te^3 te^2 te 1
    4t0^3 3t0^2 2t0 1 0
    4te^2 3te^2 2te 1 0]
C_z=inv(B_z)*A_z

z=C_z[1]*.^(t,4)+C_z[2]*.^(t,3)+C_z[3]*.^(t,2)+C_z[4]*t+C_z[5] #位置
v_z=4*C_z[1]*.^(t,3)+3*C_z[2]*.^(t,2)+2*C_z[3]*t+C_z[4] #速度
a_z=12*C_z[1]*.^(t,2)+6*C_z[2]*t+2*C_z[3]; #加速度
```

```julia
plot(t,z,st=:scatter);
```

```julia
plot(t,v_z,st=:scatter);
```

```julia
plot(t,a_z,st=:scatter);
```
