```julia
using Plots
using PyCall
using Distributions

gr()
```

```julia
a=1.0
b=1.0
D=30
μ=[i for i in 0:0.01:1]
#bernoulli
```

```julia
function bern(x,μ)
    f=μ^x * (1-μ)^(1-x)
    return f
end
```

```julia
function beta(μ,a,b)
    f=gamma(a+b)/(gamma(a)*gamma(b))*μ^(a-1) * (1-μ)^(b-1)
    
    return f
end

#gamma(a+b)/(gamma(a)*gamma(b))*0.1^(a-1)* (1-0.1)^(b-1)
```

```julia
#データ
m=Bernoulli(0.25)
x=rand(m,D)
```

```julia
N=0
#データの和を求める
x_sum=zero(eltype(x))
y=zeros(101)
for n in 1:N
    x_sum=x_sum+0
end
#事後分布を計算する
#N=0
ah=x_sum+a
bh=N-x_sum+b


for n in 1:101
    y[n]=beta(μ[n],ah,bh)
end
plot(μ,y)
```
